```python
import pandas as pd
```

```python
df=pd.read_csv('https://ironhack.school/asset-v1:IRONHACK+DAFT+201910_PAR+type@asset+block@IMDB-Movie-Data.csv')
```

```python
#on cree la colonne Avgscrore qui est la moyenne entre rating et metascore/10
df['AvgScore']=(df.Rating+df.Metascore/10)/2
```

```python
df
```

```python
#on veut creer notre propre function pour majorer si film d'horreur et minorer si fantasie
#x = score y=genre
def custom_rating(x,y):
    """x est le genre, y est le score"""
    if x.str.lower().contains("horror"):
        return min(10,y+1)
    elif x.str.lower().contains("fantasy"):
        return max(0,y-1)
    else:
        return y
```

```python
#crash because contains is only for a string
custom_rating(df.Genre,df.AvgScore)
```

```python
#on veut creer notre propre function pour majorer si film d'horreur et minorer si fantasie
#x = score y=genre
def custom_rating(x,y):
    """x est le genre, y est le score"""
    if "horror" in x.lower():
        return min(10,y+1)
    elif "fantasy" in x.lower():
        return max(0,y-1)
    else:
        return y
```

```python
#here x is the row
df.apply(lambda x: custom_rating(x.Genre, x.AvgScore),axis=1)
```

```python
# on creee la colonne correspondante
# here x is the row
df['my_score']=df.apply(lambda x: custom_rating(x.Genre, x.AvgScore),axis=1)
```

```python
#I want to locate rows with action movies
df.loc[df.Genre.str.contains('Action')]
```

```python
#maintenant on veut les lignes action qui n'ont pas plus de 2 types
# il n'y aura qu'une seule virgule
df.loc[df.Genre.str.contains('Action')&(df.Genre.str.count(',')==1)]
```

```python
#une autre methode est de ne considerer que la colonne Genre et de la scanner avec Apply
#x is then the cellule of Genre
df.loc[df.Genre.apply(lambda x: ('Action' in x)&(x.count(',')==1))];
```

```python
#same result avec map :
df.loc[df.Genre.map(lambda x: ('Action' in x)&(x.count(',')==1))]
```
